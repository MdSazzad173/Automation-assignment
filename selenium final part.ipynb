**_Libraries_** 

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

**_Inputs_** 

```python
file=input("Enter the file:\t")
wb=input("Enter the driver:\t") 
day=input("Enter the day:\t").capitalize()
```

**_Reading the file_** 

```python
#reading the file with pandas module
df=pd.read_csv(file)
df
```

**_Exicuting the task_** 

```python
#creating an object 'driver' and putting the browser drive into the object
driver = webdriver.Chrome(executable_path= wb) 

#openning the google url
driver.get ('https://www.google.com.bd')

#maximizing the browser window
driver.maximize_window ()

#this method is to hold the current line of code for the given seconds
time.sleep(2)

#finding the search box through xpath and putting it into the variable called 'main_page'
main_page=driver.find_element(By.XPATH, "/html[1]/body[1]/div[1]/div[3]/form[1]/div[1]/div[1]/div[1]/div[1]/div[2]/input[1]")

time.sleep(2)

#clicking into the search box
main_page.click()

time.sleep(2)

#for loop
for i in range(len(df[day])):
    
    #passing the key word from the csv file
    main_page.send_keys(df[day][i])
    
    time.sleep(2)
    
    #finding the html objects of the generated suggestions list
    suggestions=driver.find_elements(By.XPATH,"//div[@role='presentation']//div[@role='presentation']//ul[@role='listbox']//li")
    
    time.sleep(2)
    
    #transforming the each html object into readable text
    suggestions=[sgs.text for sgs in suggestions]
    
    #finding the maximum and minimum length of the suggestion string and putting those values into the column named "Longest Option" and "Shortest Option" 
    df['Longest Option'][i]=max(suggestions)
    df['Shortest\nOption'][i]=min(suggestions)
    main_page.clear()

#closing the browser    
driver.close()

#printing the dataframe
df
```

**_Writing into the file_** 

```python
df.to_csv(file)
```
