```python
class suggestion:
    def __init__(self, file, day):
        self.file = file
        self.day = day

    def showsuggestion(self):
        import pandas as pd
        import selenium
        from selenium import webdriver
        from selenium.webdriver.common.keys import Keys
        from selenium.webdriver.common.by import By
        import time

        driver = webdriver.Chrome(executable_path="C:\\drives\\chromedriver.exe")
        df = pd.read_csv(self.file)
        driver.get("https://www.google.com")
        side_menu = driver.find_element(By.XPATH,'//*[@id="SIvCob"]/a')
        time.sleep(2)
        side_menu.click()
#         driver.maximize_window()
#         time.sleep(2)
        main_page = driver.find_element(By.XPATH,"/html[1]/body[1]/div[1]/div[3]/form[1]/div[1]/div[1]/div[1]/div[1]/div[2]/input[1]")
#         time.sleep(2)
        main_page.click()
#         time.sleep(2)
        for i in range(len(df[day])):
            # passing the key word from the csv file
            main_page.send_keys(df[day][i])

            time.sleep(2)

            # finding the html objects of the generated suggestions list
            suggestions = driver.find_elements(By.XPATH,"//div[@role='presentation']//div[@role='presentation']//ul[@role='listbox']//li")

            time.sleep(2)

            # transforming the each html object into readable text
            suggestions = [sgs.text for sgs in suggestions]

            # finding the maximum and minimum length of the suggestion string and putting those values into the column named "Longest Option" and "Shortest Option"
            df["Longest Option"][i] = max(suggestions)
            df["Shortest\nOption"][i] = min(suggestions)
            main_page.clear()
        driver.close()
        return df


file = input("Enter the file:\t")
day = input("Enter the day:\t").capitalize()

task = suggestion(file, day)
task.showsuggestion()
```
